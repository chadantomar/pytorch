Create Tensor

1. Empty Tensor : Create empty tensor of number of rows and columns and value inside are pre assigned value at that space only .<br/> IN below example it create empty tensor of 2 rows and 3 columns

```python
a=torch.empty(2,3)
```

```python
type(a)
```

2. Zero Tensor : Create tensor of value 0 with specified no of rows and columns


```python
torch.zeros(2,3)
```

3. Ones Tensor: Create tensor of value 1 with specified no of rows and columns

```python
torch.ones(2,3)
```

4. Rand Tensor : create Tensor of random value with specified no of rows and columns

```python
torch.rand(2,3)
```

5. But every time will give differnet random number to handle such issue there is Seed concept

```python
torch.rand(2,3)
torch.manual_seed(200)
torch.rand(2,3)
```
